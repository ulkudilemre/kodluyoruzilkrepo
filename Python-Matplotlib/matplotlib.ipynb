```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Global API
Matplotlib'in varsayılan pyplot API'si, global, MATLAB tarzı bir arayüze sahiptir:

```python
x = np.arange(-10, 11)
```

```python
plt.figure(figsize = (12, 6))
plt.title("My Nice Plot")
plt.plot(x, x**2)
plt.plot(x, -1 * (x**2))
plt.show()
```

Global Api'yı kullanırken ve birden çok çizim yapıyorsanız kodd satırlarının sırası çok karmaşık olabilir.
* ilk olarak *figure* ile çizimimizin boyutunu belirtiriz.
* *title* ile çizimimize başlık veririz.
* *subplot* çizimimizin satır, sütun boyutunu belirtir ve en soldaki verdiğimiz değer, bu kodun altındaki tüm kodların hangi hücredeki çizime ait olduğunu belirtir.
* *plot* çizimi yapacağımız verileri aldığımız metottur.
* *legend* neyi görselleştirdiğimizi açıklamak için kullanılır.
* *xlabel* ve *ylabel* x ve y eksenlerini açıklak için kullanılır.



```python
plt.figure(figsize=(12, 6))
plt.title("My Nice Plot")

plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(["x^2", "Vertical Line"])
plt.xlabel("X")
plt.ylabel("X Squared")

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x**2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(["-X^2", "Horizontal Line"])

plt.xlabel("X")
plt.ylabel("X Squared")
```

Eğer 100 satırlık bir görselleştirmede hata ayıklamak gerekirse o zaman yukarıdan aşağıya neler olduğunu akılda tutmak gerekir ve bu fazlası ile zordur.

## OOP Arayüzü

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(
    x, (x**2), color="red", linewidth=3,
    marker="o", markersize=8, label="X^2")

axes.plot(x, -1 * (x**2), "b--", label="-X^2")

axes.set_xlabel("X")
axes.set_ylabel("X  Squared")

axes.set_title("My Nice Plot")

axes.legend()

fig
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x+0, linestyle='solid')
axes.plot(x, x+1, linestyle='dashed')
axes.plot(x, x+2, linestyle='dashdot')
axes.plot(x, x+3, linestyle='dotted')

axes.set_title("My Nice Plot")
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x+0, '-og', label='solid green')
axes.plot(x, x+1, '--c', label='dashed cyan')
axes.plot(x, x+2, '-.b', label='dashdot blue')
axes.plot(x, x+3, ':r', label='dotted red')

axes.set_title("My Nice Plot")

axes.legend()
```
